```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import math

script = pd.read_csv('scripts.csv')
```

```python
script["Sentiment"] = 0
columns = list(script)
scriptSentiment = pd.DataFrame(columns=columns)

for index, row in script.iterrows():
    dialogue = row['Dialogue']
    if pd.isnull(dialogue):
        continue
    testimonial = TextBlob(row['Dialogue'])
    row["Sentiment"] = testimonial.sentiment[0]    
    scriptSentiment = scriptSentiment.append(row, ignore_index=True)

    
scriptSentiment.to_csv('scriptSentiment.csv')

    
```

```python
characterColumns = ['Character', 'SEID', 'Sentiment']
characterSentiment = pd.DataFrame(columns=['Character', 'SEID', 'Sentiment'])

for episodeID, episode in scriptSentiment.groupby(['SEID']):
    for characterID, character in episode.groupby(['Character']):
        sentiment = character['Sentiment'].mean()
        thisCharacter = pd.Series([characterID, episodeID, sentiment], index=characterColumns)
        characterSentiment = characterSentiment.append(thisCharacter, ignore_index=True)

characterSentiment.to_csv('characterSentiment.csv')
```

```python
import re
def listToFreqDict(allList):
    thisFreqDF = pd.DataFrame(columns=['Word', 'Frequency'])
    
    uniqueList = list(set(allList))
    for word in uniqueList:
        thisFreq = pd.Series([word, allList.count(word)], index=['Word', 'Frequency'])
        thisFreqDF = thisFreqDF.append(thisFreq, ignore_index=True)
    return thisFreqDF


freqColumns = ['Character', 'SEID', 'Word', 'Frequency']
freqDF = pd.DataFrame(columns=freqColumns)

for episodeID, episode in script.groupby(['SEID']):
    for characterID, character in episode.groupby(['Character']):
        words = ""
        for index, row in character.iterrows():
            dialogue = row['Dialogue'] 
            if pd.isnull(dialogue):
                continue
            words += " " + dialogue             
        words = re.sub('[!@#$.(),?]', ' ', words)
        
        allList = words.split()
        
        thisFreqDF = listToFreqDict(allList)
        thisFreqDF["Character"] = characterID
        thisFreqDF["SEID"] = episodeID
        freqDF = freqDF.append(thisFreqDF)

        
freqDF.to_csv('word_frequencies.csv')
```
